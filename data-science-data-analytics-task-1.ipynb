```python
plt.hist(df['G3'])
plt.xlabel("Final Grade (G3)")
plt.ylabel("Number of Students")
plt.title("Histogram of Final Grades")
plt.show()
```

```python
plt.scatter(df['studytime'], df['G3'])
plt.xlabel("Study Time")
plt.ylabel("Final Grade (G3)")
plt.title("Study Time vs Final Grade")
plt.show()
```

```python
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.xlabel("Gender")
plt.ylabel("Average Final Grade")
plt.title("Gender-wise Average Performance")
plt.show()
```

## 1. Dataset Loading
The student performance dataset (student-mat.csv) was loaded using the Pandas library.
The dataset contains student academic and demographic information.


## 2. Data Exploration and Cleaning
The dataset was explored to understand its structure, shape, and data types.
Missing values were checked and no null values were found.
Duplicate records were also checked and removed to ensure data quality.


## 3. Dataset Inspection
The dataset consists of 395 student records and 33 attributes.
Data types include both numerical and categorical variables.


## 4. Data Analysis
The final grade (G3) was used to analyze student performance.
The average final grade was calculated.
The number of students scoring above 15 was identified.
The relationship between study time and final grade was analyzed using correlation.
Gender-wise performance was also compared.


## 5. Data Visualization
Various visualizations were created to understand the data better.
A histogram was used to show the distribution of final grades.
A scatter plot was used to analyze the relationship between study time and grades.
A bar chart was used to compare average grades between male and female students.


## 6. Conclusion
The dataset was clean and contained no missing or duplicate values.
The average final grade of students was around 10.
Study time showed a weak positive correlation with performance.
Male students performed slightly better than female students on average.
